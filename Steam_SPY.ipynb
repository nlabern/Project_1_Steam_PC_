```python
import requests
import json
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Data from the "Steam Spy API"
steam_spy_url = "https://www.steamspy.com/api.php?request=tag&tag=Early+Access"
print(requests.get(steam_spy_url))
response = requests.get(steam_spy_url).json()
print(json.dumps(response, indent=4, sort_keys=5))
```

```python
steam_spy_df = pd.DataFrame(response)
# steam_spy_df.head()
```

```python
steam_spy_df.to_csv("steamspy.csv", header=True)
steam_spy_df.head()
```

```python
# Compare developers with positive score in CSV data

#Display max 500 columns out of over 14000
pd.set_option("display.max_columns", 500)
# Set variable to read row 2 ("developers") with only unique values
dev_data = steam_spy_df.iloc[2].unique()
# Set variable to read row 5 which is called "positive".
score_pos = steam_spy_df.iloc[5]

print(dev_data)
```

```python
x_axis = np.arange(len(dev_data))
y_axis = score_pos
plt.figure(figsize=(100,25))
plt.bar(x_axis, dev_data, color="r", alpha=0.5, align="center")
plt.show()
```
